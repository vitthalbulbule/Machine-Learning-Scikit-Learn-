```python
# 1️⃣ Split data into Train and Test
# 2️⃣ Apply feature scaling ONLY on training data
# 3️⃣ Transform test data using same scaler

import pandas as pd
```

```python
df=pd.read_csv('Simple.csv')
```

```python
df.head()
```

```python
# Train Test And Split data

from sklearn.model_selection import train_test_split

x=df[['Age','Sal','Exp']]
y=df['Buy']

x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
print("x_train")
print(x_train)
print("\n x_test")
print(x_test)
```

```python
#Applying Feature Scaling On Train data and Transform Test data

from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()

x_train_scale=scalar.fit_transform(x_train)
print('\n After Feature scaling on Only Train Data')
print(x_train_scale)

print('\n After Only transform Test Data')
x_test_scaler=scalar.transform(x_test)
print(x_test_scaler)
```
