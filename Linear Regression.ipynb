```python
# Loading Data Set
import pandas as pd
df=pd.read_csv('hours_marks.csv')

df.head()
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression

model=LinearRegression()

X=df[["Hours"]]
y=df['Marks']

hours=float(input("Enter How many Hours you studied ="))

model.fit(X,y)
predict_value=model.predict([[hours]])
print(f"Based on Your Studied Hours {hours} You may score the {predict_value} Marks in the Exam")
```
