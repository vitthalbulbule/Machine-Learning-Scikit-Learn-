```python
# Feature Scaling 
# StandardScaler

import pandas as pd
df=pd.read_csv('Simple.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

new_column=['Age','Sal','Exp']
df[new_column]=scaler.fit_transform(df[new_column])
print(df)
```

```python
df=df[['Name','Gender','City','Adult','Age','Sal','Exp','Buy']]
df.head()
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['adult_encode']=le.fit_transform(df['Adult'])
df
```

```python
# Normalization (Min-Max Scaler) Ra

from sklearn.preprocessing import MinMaxScaler
mscaler=MinMaxScaler()

df[new_column]=mscaler.fit_transform(df[new_column])
df
```
